```python
##Rsolution dun systme dquations linaires - numpy.linalg.solve()
Pour rsoudre le systme dquations linaires 3 * x0 + x1 = 9 et x0 + 2 * x1 = 8 :###
```

```python
a = np.array([[3,1], [1,2]])

b = np.array([9,8])

x = np.linalg.solve(a, b)
print(x)
```

```python
##Pour vrifier que la solution est correcte :##
```

```python
np.allclose(np.dot(a, x), b)
```
